```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([60, 90, 70,85 ,80,65,75,60,75,70])
y = np.array([70,95, 70, 75, 90,75,75,60,80,65])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')


plt.title("Scatter Plot of the Correlation of Exam Scores and Course Grades")
plt.xlabel("Exam Score")
plt.ylabel("Course Grade")
plt.show()
```

Interpretation:

The scatter plot demonstrates a positive correlation between exam scores and course grades, as indicated by the upward-sloping red trend line. This suggests that higher exam scores are generally associated with higher course grades.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2,100)
y = x**2
fig=  plt.figure(figsize = (10,5))
plt.plot(x,y)
plt.show()
```

```python

#Generate data
x = np.array([0.5, 1, 1.3, 1.6, 2, 2.2, 2.5, 3.1, 3.9, 4.4])
y = np.array([1.9, 1.5, 1.2, 1, 0.8, 0.78, 0.65, 0.46, 0.3, 0.23])

#Calculate the slop and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

#Calculate the predicted values of y using the regression line
y_pred = slope * x + intercept

#Plot the original data points and the regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title("Scatter Plot of the Correlation of Samples and Data Sets")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Interpretation:


The scatter plot shows a negative correlation between the variables

x and

y, indicating that as

x increases,

y decreases. The red line represents the line of best fit, capturing the downward trend in the data.
