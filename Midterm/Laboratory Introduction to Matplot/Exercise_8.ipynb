Matplot Labels and Title - You can use the xlabel() and ylabel()

```python
xpoints = np.array([1,2,6,8]) #plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints)

plt.title("Distance Travelled Everyday", color='red')
plt.xlabel("Distance in km") #labels
plt.ylabel("Speed in km/s")
plt.show()
```

Matplot Grid

```python
xpoints = np.array([1,2,6,8]) #plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints)#plot() is used to draw points
plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("Speed in km/s")
plt.grid(axis='y')
plt.show()
```

Show Multiple Plots

```python
#plot 1
xpoints = np.array([1,2,6,8]) #plotting x and y points
ypoints = np.array([3,8,1,10])
plt.subplot(1,2,1)

plt.plot(xpoints, ypoints)#plot() is used to draw points

#plot 2
xpoints = np.array([1,2,6,8]) #plotting x and y points
ypoints = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(xpoints, ypoints)#plot() is used to draw points

plt.grid(axis='x', linestyle='--', linewidth=0.50)
plt.show()
```

Matplotlib Pie Chart - You can use pie() to draw pie chart

```python
g = np.array([35, 25, 25, 15])
grader = ['Grade 1=35%', 'Grade 2=25%', 'Grade 3=25%', 'Grade 4=15%']
plt.pie(g,labels=grader)
#plt.show()
plt.legend(title = "Grader",labels=grader)
```

Font Style

```python
g = np.array([35, 25, 25, 15])
grader = ['Grade 1=35%', 'Grade 2=25%', 'Grade 3=25%', 'Grade 4=15%']
plt.pie(g,labels=grader)
#plt.show()
plt.legend(title="Grader", labels=grader, title_fontproperties={'style': 'italic'})
```

1. Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x,y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x,y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

3. Histogram

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.show()
```

4. Bar Chart


```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

5. Box plot


```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 + [3]*10]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show
```

6. Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black') # bins covers a specific range of values
plt.title('Distribution of Gym Member Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt

plt.scatter(df['Avg_BPM'], df['Fat_Percentage'], color='blue', alpha=0.5)

plt.title('Average Heart Rate (Avg_BPM) vs. Fat Percentage', fontsize=14)
plt.xlabel('Average BPM', fontsize=12)
plt.ylabel('Fat Percentage', fontsize=12)

plt.grid(True)
plt.show()
```

```python
#Bar Chart
import matplotlib.pyplot as plt

workout_counts = df['Workout_Type'].value_counts()

workout_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Each Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right') # rotation of the variables
plt.tight_layout() #spacing of bar
plt.show()
```
