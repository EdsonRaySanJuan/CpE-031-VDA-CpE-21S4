1. Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x,y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x,y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

3. Histogram

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.show()
```

4. Bar Chart


```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

5. Box plot


```python
import matplotlib.pyplot as plt

data = [[1.5]*10 + [2]*10 + [3]*10]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show
```

6. Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```

```python
#Pie Graph
labels = df['Gender'].unique()

plt.pie(df['Gender'].value_counts(), labels=labels, autopct='%1.1f%%', startangle=90)  # Use value_counts() for frequency and assign labels
plt.title('Gender Distribution')
plt.show()
```

```python
# Multiple Plots
gym_data = pd.read_csv('gym_members_exercise_tracking.csv')

avg_data_by_gender = gym_data.groupby('Gender')[['Session_Duration (hours)', 'Calories_Burned']].mean()

# Prepare data for plotting
genders = avg_data_by_gender.index
avg_session_duration = avg_data_by_gender['Session_Duration (hours)']
avg_calories_burned = avg_data_by_gender['Calories_Burned']

# Set up subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot average session duration
ax[0].bar(genders, avg_session_duration, color=['skyblue', 'salmon'])
ax[0].set_title('Average Session Duration by Gender')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Average Duration (hours)')

# Plot average calories burned
ax[1].bar(genders, avg_calories_burned, color=['skyblue', 'salmon'])
ax[1].set_title('Average Calories Burned by Gender')
ax[1].set_xlabel('Gender')
ax[1].set_ylabel('Average Calories Burned')

plt.tight_layout()
plt.show()
```

```python
# Line Plot
# Define age groups
bins = [18, 25, 35, 45, 55, 65, 75]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74']
gym_data['Age_Group'] = pd.cut(gym_data['Age'], bins=bins, labels=labels)

# Calculate the average resting BPM for each age group
avg_resting_bpm_by_age_group = gym_data.groupby('Age_Group')['Resting_BPM'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(avg_resting_bpm_by_age_group.index, avg_resting_bpm_by_age_group.values, marker='o', linestyle='-', color='purple')
plt.title('Average Resting BPM by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Resting BPM')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```
