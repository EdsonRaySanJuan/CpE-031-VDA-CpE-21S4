```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = '/content/heart_2020_cleaned.csv'

df = pd.read_csv(file_path)
df.head()
```

Question 1:
Which gender has the highest count of diabetes across different age categories?

```python
#Diabetic
diabetic_gender_age_counts = df.groupby(['AgeCategory', 'Diabetic', 'Sex']).size().unstack(fill_value=0)

# Plotting the data
plt.figure(figsize=(12, 8))  # Adjust figure size as needed

# Loop through each age category
for age_category in diabetic_gender_age_counts.index.get_level_values(0).unique():
    # Extract data for the current age category
    age_data = diabetic_gender_age_counts.loc[age_category]

    # Create a new figure for each age category
    plt.figure(figsize=(8, 6))

    # Plot the data for the current age category
    age_data.plot(kind='bar', color=['skyblue', 'salmon'], label=age_category)

    # Customize plot elements
    plt.title(f'Distribution of Diabetic Status by Gender for {age_category}')
    plt.xlabel('Diabetic Status')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Gender')
    plt.tight_layout()

    # Show the plot for the current age category
    plt.show()
```

Observation:
Based on the analysis of the data for diabetic across different age categories for both males and females, it is evident that diabetes increases with age for both genders. Overall, the data indicates majority of female and male said No that they have diabetes.

Question 2:

What are the smoking rates among males and females in different age categories?

```python
#smoking
ge_category_distribution = df.groupby(['Smoking', 'Sex'])['AgeCategory'].value_counts().unstack()


# Plot the data
ge_category_distribution.plot(kind='bar', stacked=True)  # Changed age_category_distribution to ge_category_distribution
plt.title('Age Category Distribution within Smoking and Gender')
plt.xlabel('Smoking Status and Gender')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Age Category')
plt.show()
```

Observatiion:
Based on the analysis of the data that for smoking habits across different age categories for both males and females, it is evident that smoking prevalence varies significantly with age and gender. Overall, female shows a higher rate of No answer compared to male
